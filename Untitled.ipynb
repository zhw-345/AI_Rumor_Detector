```python
def event_aware_retriever(query, event_name=None, k=5):
    """优先检索同一事件的案例，不足时补充其他事件"""
    
    # 先查同一事件
    if event_name:
        same_event_filter = {"event": event_name}
        same_event_docs = case_db.similarity_search(
            query, 
            k=k, 
            filter=same_event_filter
        )
        if len(same_event_docs) >= k:
            return same_event_docs
    
    # 不足时全局检索
    return case_db.similarity_search(query, k=k)
```

```python
import os
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"#Hugging Face 连接不稳定，使用镜像源

import pandas as pd
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_chroma import Chroma
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_huggingface import HuggingFaceEmbeddings

# ==================== 配置区：提取自你的 API 调用函数 ====================
API_KEY = "sk-Th_ZuQs31rz8M72BwOViKw"           # 替换为你的真实 Key
BASE_URL = "https://models.sjtu.edu.cn/api/v1"  # 交大模型平台接口
CHAT_MODEL = "deepseek-chat"                    # 对话模型
EMBEDDING_MODEL = "sentence-transformers/all-mpnet-base-v2"      # Embedding 模型（英文sentence-transformers/all-mpnet-base-v2；中文BAAI/bge-small-zh-v1.5）
TEMPERATURE = 0.5
MAX_TOKENS = 1024
TIMEOUT = 180          # 单次请求最多等180秒
MAX_RETRIES = 5        # 自动重试 5 次
'''
# 1. 加载文档
filepath="./data/train.csv"
all_docs = pd.read_csv(filepath)
docs = []
for i, row in all_docs.iterrows():
    content = '\n'.join([f'{col}:{row[col]}' for col in ["text","label"] if pd.notna(row[col])])
    docs.append(Document(
            page_content=content,
            metadata={"source": filepath, "row_index": i}
    ))
print(f"加载了 {len(docs)} 个文档片段")

# 2. 分块（CSV 每行通常较短，参数可按实际内容调整）
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", "。", "，", " ", ""],
)
chunks = splitter.split_documents(docs)
print(f"分块后: {len(chunks)} 个块")
'''
embeddings = HuggingFaceEmbeddings(
    model_name=EMBEDDING_MODEL,
    model_kwargs={"device": "cuda"},
    encode_kwargs={"normalize_embeddings": True}
)
print("嵌入模型加载完成")

# 4. 存入向量库
vectorstore = Chroma(
            persist_directory="./chroma_pheme_cases",
            embedding_function=embeddings
        )
print("案例库加载成功")

# 5. 检索器
retriever = vectorstore.as_retriever(
        search_type="mmr",
        search_kwargs={
            "k": 5,
            "fetch_k": min(5 * 4, 50),  # 候选池
            "lambda_mult": 0.7
        }
    )
'''retriever = vectorstore.as_retriever(
    search_type="similarity",#mmr:平衡相关性与多样性,增加fetch_k
    search_kwargs={"k": 5}
)'''

# 6. LLM —— 核心：提取你的 API 信息，指向交大平台
llm = ChatOpenAI(
    model=CHAT_MODEL,
    base_url=BASE_URL,
    api_key=API_KEY,
    temperature=TEMPERATURE,
    max_tokens=MAX_TOKENS,
    timeout=TIMEOUT,
    max_retries=MAX_RETRIES,
    # 如果你需要完全复刻你原代码的"4xx(非429)直接放弃"逻辑，可额外配置：
    # http_client=httpx.Client(timeout=TIMEOUT),  # 需要 import httpx
)

# 7. Prompt
prompt = ChatPromptTemplate.from_template("""
Answer the user's question based on the following reference materials.
If the materials contain no information related to the question at all, explicitly state: 
"I cannot answer this based on the available information."

However, if the materials do contain relevant information, answer based on what is provided, 
even if the information is incomplete. Do not add any facts not present in the materials.

Reference Materials:
{context}

User Question:
{question}

Answer:
""")

'''# 8. 组装 RAG 链（LCEL 语法，LangChain 0.2.x+ 标准写法）
rag_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
)'''
def format_cases(docs):
    """格式化案例，包含 metadata"""
    formatted = []
    for doc in docs:
        meta = doc.metadata
        block = (
            f"[Case ID: {meta.get('case_id')} | "
            f"Event: {meta.get('event')} | "
            f"Label: {meta.get('label')}] |"
            f"Reactions: {meta.get('reactions')}] |"
            f"Doubt: {meta.get('doubt')}] |"
            f"Support: {meta.get('support')}] |"
            f"Text: {doc.page_content}"
        )
        formatted.append(block)
    return "\n\n---\n\n".join(formatted)

def explain_with_pheme(target_text: str, predicted_label: str, confidence: float, retriever):
    similar_cases = retriever.invoke(target_text)
    
    chain_input = {
        "target_text": target_text,
        "predicted_label": predicted_label,
        "confidence": confidence,
        "similar_cases": format_cases(similar_cases)
    }
    
    messages = prompt.format_messages(**chain_input)
    response = llm.invoke(messages)
    print(response.response_metadata.get("token_usage", {}))
    return response.content, similar_cases


# 9. 运行
if __name__ == "__main__":
    text,label,conf = input().split()
    explanation, sources = explain_with_pheme(text, label, conf, retriever)
    print(f"Retrieved {len(sources)} cases")
    print("Explanation:")
    print(explanation)
    print("-" * 40)
```
